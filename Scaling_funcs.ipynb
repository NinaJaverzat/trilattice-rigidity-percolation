```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
'''
RP critical exponents from D. J. Jacobs and M. F. Thorpe Phys. Rev. E 53, 3682 (1996)
'''

nu = 1.21
beta = 0.175
pc = 0.66
Df = 1.86
gamma = nu*(2*Df-2)
```

```python
num=1 #number of your run
DIR='./res/'
#Percolation proba

p5=np.loadtxt(DIR+f'PP_L32_T500_num{num}.txt')
p6=np.loadtxt(DIR+f'PP_L64_T200_num{num}.txt')
p7=np.loadtxt(DIR+f'PP_L128_T100_num{num}.txt')
p8=np.loadtxt(DIR+f'PP_L256_T100_num{num}.txt')


#Percolation strength

pinf5=np.loadtxt(DIR+f'ROP_L32_T500_num{num}.txt')
pinf6=np.loadtxt(DIR+f'ROP_L64_T200_num{num}.txt')
pinf7=np.loadtxt(DIR+f'ROP_L128_T100_num{num}.txt')
pinf8=np.loadtxt(DIR+f'ROP_L256_T100_num{num}.txt')

#Susceptibility
chi5=np.loadtxt(DIR+f'CHI_L32_T500_num{num}.txt')
chi6=np.loadtxt(DIR+f'CHI_L64_T200_num{num}.txt')
chi7=np.loadtxt(DIR+f'CHI_L128_T100_num{num}.txt')
chi8=np.loadtxt(DIR+f'CHI_L256_T100_num{num}.txt')
```

```python
'''
Usual probability to percolate:
P = P_both + 1/2(P_vertical + P_horizontal)
'''

def PP(tab):
    return tab[:,3]+.5*(tab[:,1]+tab[:,2])
```

```python
plt.figure(figsize=(20, 8))

plt.subplot(1, 3, 1)
plt.errorbar(p5[:,0],p5[:,1], yerr = np.sqrt(p5[:,1]*(1-p5[:,1])), fmt ='o')
plt.errorbar(p6[:,0],p6[:,1], yerr = np.sqrt(p6[:,1]*(1-p6[:,1])), fmt ='o')
plt.errorbar(p7[:,0],p7[:,1], yerr = np.sqrt(p7[:,1]*(1-p7[:,1])), fmt ='o')
plt.errorbar(p8[:,0],p8[:,1], yerr = np.sqrt(p8[:,1]*(1-p8[:,1])), fmt ='o')
plt.xlabel(r'$L$',fontsize=20)
plt.ylabel(r'$P_v$',fontsize=20)

plt.subplot(1, 3, 2)
plt.errorbar(p5[:,0],p5[:,2], yerr = np.sqrt(p5[:,2]*(1-p5[:,2])), fmt ='o')
plt.errorbar(p6[:,0],p6[:,2], yerr = np.sqrt(p6[:,2]*(1-p6[:,2])), fmt ='o')
plt.errorbar(p7[:,0],p7[:,2], yerr = np.sqrt(p7[:,2]*(1-p7[:,2])), fmt ='o')
plt.errorbar(p8[:,0],p8[:,2], yerr = np.sqrt(p8[:,2]*(1-p8[:,2])), fmt ='o')
plt.xlabel(r'$L$',fontsize=20)
plt.ylabel(r'$P_h$',fontsize=20)

plt.subplot(1, 3, 3)
plt.errorbar(p5[:,0],p5[:,3], yerr = np.sqrt(p5[:,3]*(1-p5[:,3])), fmt ='o')
plt.errorbar(p6[:,0],p6[:,3], yerr = np.sqrt(p6[:,3]*(1-p6[:,3])), fmt ='o')
plt.errorbar(p7[:,0],p7[:,3], yerr = np.sqrt(p7[:,3]*(1-p7[:,3])), fmt ='o')
plt.errorbar(p8[:,0],p8[:,3], yerr = np.sqrt(p8[:,3]*(1-p8[:,3])), fmt ='o')
plt.xlabel(r'$L$',fontsize=20)
plt.ylabel(r'$P_b$',fontsize=20)


plt.show()

plt.figure(figsize=(20, 8))

plt.subplot(1, 2, 1)
plt.plot(p5[:,0],PP(p5),marker='o')
plt.plot(p6[:,0],PP(p6),marker='o')
plt.plot(p7[:,0],PP(p7),marker='o')
plt.plot(p8[:,0],PP(p8),marker='o')
plt.xlabel(r'$L$',fontsize=20)
plt.ylabel(r'$P_b+1/2(P_h+P_v)$',fontsize=20)

plt.subplot(1, 2, 2)
plt.plot(32**(1/nu)*(p5[:,0]-pc),PP(p5),marker='o')
plt.plot(64**(1/nu)*(p6[:,0]-pc),PP(p6),marker='o')
plt.plot(128**(1/nu)*(p7[:,0]-pc),PP(p7),marker='o')
plt.plot(256**(1/nu)*(p8[:,0]-pc),PP(p8),marker='o')
plt.xlabel(r'$L^{1/\nu}(p-p_c)$',fontsize=20)
plt.ylabel(r'$P_b+1/2(P_h+P_v)$',fontsize=20)

plt.show()
```

```python
#plt.rcParams['text.usetex'] = True

plt.figure(figsize=(20, 8))

plt.subplot(1, 2, 1)
plt.errorbar(pinf5[:,0],pinf5[:,1], yerr = np.sqrt(pinf5[:,2]-pinf5[:,1]**2), fmt ='o')
plt.errorbar(pinf6[:,0],pinf6[:,1], yerr = np.sqrt(pinf6[:,2]-pinf6[:,1]**2), fmt ='o')
plt.errorbar(pinf7[:,0],pinf7[:,1], yerr = np.sqrt(pinf7[:,2]-pinf7[:,1]**2), fmt ='o')
plt.errorbar(pinf8[:,0],pinf8[:,1], yerr = np.sqrt(pinf8[:,2]-pinf8[:,1]**2), fmt ='o')
plt.xlabel(r'$L$',fontsize=20)
plt.ylabel(r'$P_\infty$',fontsize=20)

plt.subplot(1, 2, 2)
plt.errorbar(32**(1/nu)*(pinf5[:,0]-pc),32**(beta/nu)*pinf5[:,1], yerr = np.sqrt(pinf5[:,2]-pinf5[:,1]**2), fmt ='o')
plt.errorbar(64**(1/nu)*(pinf6[:,0]-pc),64**(beta/nu)*pinf6[:,1], yerr = np.sqrt(pinf6[:,2]-pinf6[:,1]**2), fmt ='o')
plt.errorbar(128**(1/nu)*(pinf7[:,0]-pc),128**(beta/nu)*pinf7[:,1], yerr = np.sqrt(pinf7[:,2]-pinf7[:,1]**2), fmt ='o')
plt.errorbar(256**(1/nu)*(pinf8[:,0]-pc),256**(beta/nu)*pinf8[:,1], yerr = np.sqrt(pinf8[:,2]-pinf8[:,1]**2), fmt ='o')
plt.xlabel(r'$L^{1/\nu}(p-p_c)$',fontsize=20)
plt.ylabel(r'$L^{\beta/\nu}\,P_\infty$',fontsize=20)
#plt.tight_layout(4)
plt.show()
```
